# SPKIR CG data deep dive 1.3

In this notebook we display OOI SPKIR data stored in the kdata drive for certain deployments that may need additional annotations. Ref Des: CP04OSSM-RID26-08-SPKIRB000

```python
import numpy as np
import pandas as pd
import xarray as xr
import os
import matplotlib.pyplot as plt
import matplotlib.colors as colors
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from scipy.stats import norm
from scipy.signal import find_peaks

from ooi_data_explorations.common import list_methods, list_streams, get_annotations, add_annotation_qc_flags, load_kdata, get_vocabulary, update_dataset
from ooi_data_explorations.uncabled.process_spkir import spkir_datalogger
from ooinet.M2M import get_deployments
from qartod_testing.data_deep_dive import nanfill_time_gaps, check_chla_swr, compare_spkir_to_ctdbp
```

```python
# Set parameters for data request 
refdes = 'CP04OSSM-RID26-08-SPKIRB000'
site, node, sensor = refdes.split('-', 2)
```

```python
# Set parameter for SPKIR comparison with FLORT Chl-a estimate
flort_node = 'RID27'
```

```python
# Set parameter for SPKIR internal temperature comparison with co-located CTDBP sea water temperature
ctdbp_node = 'RID27'
ctdbp_sensor = '03-CTDBPC000'
```

### Check data delivery method, deployments, and annotations

```python
# Create dict of methods and associated data streams
methods = list_methods(site, node, sensor)

streams = {}
for method in methods:
    streams[method] = list_streams(site, node, sensor, method)

streams
```

```python
# View deployment information
deployments = get_deployments(refdes)
deployments
```

```python
## View annotations
annotations = get_annotations(site, node, sensor)
annotations = pd.DataFrame(annotations)

if not annotations.empty:
    annotations = annotations.drop(columns=['@class'])
    annotations['beginDate'] = pd.to_datetime(annotations.beginDT, unit='ms').dt.strftime('%Y-%m-%dT%H:%M:%S')
    annotations['endDate'] = pd.to_datetime(annotations.endDT, unit='ms').dt.strftime('%Y-%m-%dT%H:%M:%S')

annotations
# annotations['annotation'][5]

# Note: now I have moved the lines that add a human-readable date time up here where the annotations are displayed, however I'm not sure if the begin and endDT columns still need 
# to exist. I would like to remove them because they are just taking up space and don't add anything to the dataFrame displayed below.
```

### Deployment 5

Internal temperature steeply increases ~4 degC within a day on 2017-03-19

The SPKIR internal temperature turned out to be very similar to the co-located CTDBP sea water temperature except for an offset and some of the day-to-day changes. The random prolonged drop in temperature that I was concerned about is also reflected in the sea water temperature time series. I guess I don't need an annotation for this one.

I wasn't going to do anything else for this one, but I'm curious what all the other sensors were seeing at the same time, because I still find the sudden drop in water temperature in the CTD record odd. Was a different mass of water suddenly advected past the sensors? Did the winds change and mess with the normal dynamics of the area?

```python
# Set method and stream strings
method = methods[0]
stream = streams[method][0]
deploy = 5
```

```python
# Request annotations and load data from the local kdata directory
annotations = get_annotations(site, node, sensor)
annotations = pd.DataFrame(annotations)
spkir = load_kdata(site, node, sensor, method, stream, ('*deployment%04d*SPKIR*.nc' % deploy))
spkir = spkir_datalogger(spkir, burst=True)
```

```python
# Add qc flags for annotations retrieved above
if not annotations.empty:
    # create an annotation-based quality flag
    spkir = add_annotation_qc_flags(spkir, annotations)

    # clean-up the data, removing values that were marked as fail either from the quality checks or in the
    # annotations
    spkir = spkir.where(spkir.rollup_annotations_qc_results < 4) # flags from annotations
    # spkir = spkir.where(spkir. # not sure which summary_qc variables are in here that I should use for this part -- there isn't such a variable in these datasets
```

```python
spkir = nanfill_time_gaps(spkir)
```

```python
# Check structure of edited data and contents
vocab = get_vocabulary(site, node, sensor)[0]
spkir = update_dataset(spkir, vocab['maxdepth'])
spkir['time'] = pd.to_datetime(spkir.time, unit='s')
spkir
```

```python
# First plot edited data to check whether inserting nans over gaps in time coverage is necessary
fig, ax = plt.subplots(1,1,figsize=(15,4))
for var in spkir.variables:
    if "downwelling_irradiance" in var:
        spkir[var].plot(ax=ax, label=spkir[var].radiation_wavelength)
ax.set_ylabel('Downwelling Spectral Irradiance [uW cm$^{-2}$ nm$^{-1}$]')
plt.legend()
```

```python
#  Checking internal temperature, supply voltage, and analog rail voltage during this time
fig, ax = plt.subplots(3,1, sharex=True, figsize=(15,8))
spkir.internal_temperature[0].plot.line(ax=ax[0])
spkir.input_voltage[0].plot.line(ax=ax[1])
spkir.analog_rail_voltage[0].plot.line(ax=ax[2])
```

```python
# Compare SPKIR internal temperature time series to sea water temperature
# to check that the instrument is performing as expected
ctdbp, fig, ax = compare_spkir_to_ctdbp(spkir, site, deploy, ctdbp_node, ctdbp_sensor)
```

```python
# Find start date for annotation
# We'll start the annotation from the day after the lowest
# local max in Aug since this is where the SPKIR signal declines 
# compared to the SWR
spkir_aug_555 = spkir.sel(
    time=slice("1 Aug 2021",
               "31 Aug 2021"))["downwelling_irradiance_555"].squeeze(
    drop=True)
spkir_aug_555 = spkir_aug_555.where(spkir_aug_555>=0, drop=True)
local_max_aug, peak_properties = find_peaks(spkir_aug_555.data,
                                           distance=48)
start = spkir_aug_555.where(
    spkir_aug_555==min(spkir_aug_555[local_max_aug]),
    drop=True)
print(start.time)
```

```python
# Find minimum after start of reduced SPKIR signal before next peak
time_check = spkir_aug_555.sel(time=slice("2021-08-04T14:15:00", "2021-08-05T15:30:00"))
print(time_check.time.where(time_check==time_check.min(),drop=True))
```

```python
# Find end date for annotation from the last peak 
# in Sept 2021 below ~40 uW/cm2/nm
spkir_sept_555 = spkir.sel(
    time=slice("1 Sept 2021",
               "30 Sept 2021"))["downwelling_irradiance_555"].squeeze(
    drop=True)
spkir_sept_555 = spkir_sept_555.where(spkir_sept_555>=0, drop=True)
local_max_sept, peak_properties = find_peaks(spkir_sept_555.data,
                                           distance=48)
fig = plt.figure()
spkir_sept_555.plot()
spkir_sept_555[local_max_sept].where(spkir_sept_555[local_max_sept]<40,
                                     drop=True).plot.scatter(c="black")
print(spkir_sept_555[local_max_sept].where(spkir_sept_555[local_max_sept]<40,
                                     drop=True)[-1].time)
```

```python
# Use time of minimum between last peak in annotation
# and next peak for end dt of annotation
t0 = spkir_sept_555[local_max_sept].where(spkir_sept_555[local_max_sept]<40,
                                     drop=True)[-1].time
t1 = spkir_sept_555[local_max_sept].sel(time=slice("2021-09-25"))[-1].time
minval = spkir_sept_555.sel(time=slice(t0, t1)).min()
print(spkir_sept_555.where(spkir_sept_555.sel(time=slice(t0, t1))==minval, drop=True))
```
