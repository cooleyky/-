### Deployment 2

Potential biofouling starting around the 11-month mark of deployment.

Based on the Chl-a estimate and the downwelling SWR, I would start to suspect that the instrument was biofouled from the last gap in the data to the end of the deployment. The SPKIR signal doesn't recover after the austral winter like the SWR does, even though they declined similarly from summer to winter. The Chl-a estimate from the FLORT also shows how a large bloom at the start of the deployment affected the SPKIR signal, after which the SPKIR recovers. Throughout the rest of the deployment we do not see any more significant spikes in the Chl-a that would interfere with light reaching the SPKIR. This data should probably be flagged as suspect.

```python
# Set method and stream strings
method = methods[0]
stream = streams[method][0]
deploy = 2
```

```python
# load data from the local kdata directory
spkir = load_kdata(site, node, sensor, method, stream, ('*deployment%04d*SPKIR*.nc' % deploy))
spkir = spkir_datalogger(spkir, burst=True)
```

```python
# Add qc flags for annotations retrieved above
if not annotations.empty:
    # create an annotation-based quality flag
    spkir = add_annotation_qc_flags(spkir, annotations)

    # clean-up the data, removing values that were marked as fail either from the quality checks or in the
    # annotations
    spkir = spkir.where(spkir.rollup_annotations_qc_results < 4) # flags from annotations
    # spkir = spkir.where(spkir. # not sure which summary_qc variables are in here that I should use for this part -- there isn't such a variable in these datasets
```

```python
spkir = nanfill_time_gaps(spkir)
```

```python
# Check structure of edited data and contents
vocab = get_vocabulary(site, node, sensor)[0]
spkir = update_dataset(spkir, vocab['maxdepth'])
spkir['time'] = pd.to_datetime(spkir.time, unit='s')
spkir
```

```python
# First plot edited data to check whether inserting nans over gaps in time coverage is necessary
fig, ax = plt.subplots(1,1,figsize=(15,4))
for var in spkir.variables:
    if "downwelling_irradiance" in var:
        spkir[var].plot(ax=ax, label=spkir[var].radiation_wavelength)
ax.set_ylabel('Downwelling Spectral Irradiance [uW cm$^{-2}$ nm$^{-1}$]')
plt.legend()
```

```python
#  Checking internal temperature, supply voltage, and analog rail voltage during this time
fig, ax = plt.subplots(3,1, sharex=True, figsize=(15,8))
spkir.internal_temperature[0].plot.line(ax=ax[0])
spkir.input_voltage[0].plot.line(ax=ax[1])
spkir.analog_rail_voltage[0].plot.line(ax=ax[2])
```

```python
# Plot irradiance against SWR from the surface buoy METBK, co-located FLORT 
# site, node and sensor of the SPKIR are already defined as such
# site will be the same also for METBK and FLORT sensors. While the
# FLORT is also on the NSIF, its data goes through a different DCL such
# that the node will be different as well.
metbk, flort, fig, ax = check_chla_swr(spkir, site, deploy, flort_node)

ax[0].set_xlim([np.datetime64("2016-09-30"),np.datetime64("2016-11-14")])
fig.show()
```

```python
# Plot Oct 2016 to find good start date for annotation
# We're look for a point where downwelling irradiance stops increasing along seasonal trend
fig, ax = plt.subplots(1,1,figsize=(15,4))
for var in spkir.variables:
    if "downwelling_irradiance" in var:
        spkir_plt = spkir[var].sel(time=slice(np.datetime64("2016-10-28"), np.datetime64("2016-11-05"))).squeeze()
        spkir_pk, _ = find_peaks(spkir_plt, distance=48)
        spkir_plt.plot(ax=ax, label=spkir[var].radiation_wavelength)
        spkir_plt[spkir_pk].plot.scatter(ax=ax) 
ax.set_ylabel('Downwelling Spectral Irradiance [uW cm$^{-2}$ nm$^{-1}$]')
```

### Deployment 3

Potential biofouling starting around the 11-month mark of deployment. 

While we probably won't annotate this, I also think we should probably manually exclude the one or a few data points that were recorded after a long gap in the data.

The Chl-a estimate is also a little funky towards the end of the deployment, but I think we can still get an idea of what the relative influence of chlorophyll blooms on the downwelling spectral irradiance should be. After a period of no blooms and decreased downwelling SWR in the winter, the SPKIR doesn't recover until after a large bloom in the spring, but the SPKIR signal still doesn't reach the a similar magnitude as an earlier period when the SWR signal was at the same magnitude. This diminished magnitude starts just before October 2017. The January-March 2018 SPKIR data looks to be a similar magnitude to the same time of year before the suspected biofouled data. Beyond that brief period, I suspect that the data from Oct 2017 through the end of the deployment could be biofouled.

### Histograms of downwelling irradiance measurements

```python
# histogram of downwelling irradiance @ 412 nm variable
irrad412 = spkir['downwelling_irradiance_412'][0]
# irrad412 = spkir['downwelling_irradiance_412'][0].sel(time=slice('2017-07-01'))
mu412 = irrad412.mean()
sig412 = irrad412.std()

plt.figure(figsize=(15,4))
c, x, h = irrad412.plot.hist(bins=30)
norm412 = sum(c)*norm.pdf(x, mu412, sig412)
plt.plot(x, norm412)
```

```python
# concat the different irradiance variables for histogram of all irradiance values
irradAll = xr.DataArray()
for var in spkir.variables:
    if "downwelling_irradiance" in var:
        irradAll = xr.concat([irradAll, spkir[var][0]], dim='channel')
irradAll
```

```python
# histogram using all irradiance values
radAll= irradAll
muAll = radAll.mean()
sigAll = radAll.std()

plt.figure(figsize=(15,4))
c, x, h = radAll.plot.hist(bins=40)
normAll = sum(c)*norm.pdf(x, muAll, sigAll)
plt.plot(x, normAll)
```
